## Feature engineering

```python


df_encoded = df_clean.copy()
df_encoded = df_encoded.drop('customerID', axis=1)
```

```python
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 
                   'PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                   'TechSupport', 'StreamingTV', 'StreamingMovies',
                   'Contract', 'PaperlessBilling', 'PaymentMethod']
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

# Encode target variable
df_encoded['Churn'] = df_encoded['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df_encoded
```

## Prepare Data for Modeling

```python
%pip install imbalanced-learn
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve,recall_score,f1_score,precision_score
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')
sns.set_palette("husl")
%matplotlib inline
```

```python
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.value_counts()
```

```python
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

## Model building pipelines

```python
pipelines = {
    'RandomForestClassifier': Pipeline([
        ('scaler', StandardScaler()),
        ('model',RandomForestClassifier(random_state=42))
    ]),

    'LogisticRegression' : Pipeline([
        ('Scaler',StandardScaler()),
        ('model', LogisticRegression(random_state=42))
    ]),

    'SVM' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', SVC(random_state=42, probability=True))
    ]),

    'GradientBoostingClassifier' : Pipeline([
        ('scaler',StandardScaler()),
        ('model', GradientBoostingClassifier(random_state=42))
    ])


}

# Hyperparameter grids
param_grids = {
    'RandomForestClassifier': {
        'model__n_estimators': [100, 200],
        'model__max_depth': [10, 20, None],
        'model__min_samples_split': [2, 5],
        'model__class_weight': ['balanced', None]
    },
    'GradientBoostingClassifier': {
        'model__n_estimators': [100, 200],
        'model__learning_rate': [0.1, 0.05],
        'model__max_depth': [3, 5]
    },
    'LogisticRegression': {
        'model__C': [0.1, 1, 10],
        'model__solver': ['liblinear', 'lbfgs'],
        'model__class_weight': ['balanced', None]
    },
    'SVM': {
        'model__C': [0.1, 1, 10],
        'model__kernel': ['rbf', 'linear'],
        'model__class_weight': ['balanced', None]
    }
}

```

```python
print("Trainig and evaluating")

results = {}
best_models = {}
cv_results = {}

for name in pipelines.keys():
    grid_search = GridSearchCV(
        pipelines[name],
        param_grids[name],
        cv=5,
        scoring='roc_auc',
        n_jobs=-1,
        verbose=0
    )
    grid_search.fit(X_train_balanced,y_train_balanced)
    # save model
    best_models[name] = grid_search.best_estimator_
    results[name] = grid_search.best_score_
    cv_results[name] = grid_search.cv_results_
```

```python
print(f"Best AUC for {name}: {grid_search.best_score_}")
print(f"Best Params: {grid_search.best_params_}\n")
print(f"Best model {name}:\n")
```

```python
y_pred = best_models[name].predict(X_test)
y_pred_proba = best_models[name].predict_proba(X_test)[:, 1]
    
```

```python
  # Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
auc_score = roc_auc_score(y_test, y_pred_proba)
    
```

```python
print(accuracy, auc_score)
```

```python
cv_score = cross_val_score(best_models[name],X_train_balanced,y_train_balanced,cv = 5,scoring = 'roc_auc')
```

```python
cv_score
```

```python
class_report = classification_report(y_test,y_pred,output_dict=True)
```

```python
results[name] = {
        'best_params': grid_search.best_params_,
        'test_accuracy': accuracy,
        'test_auc': auc_score,
        'cv_mean': cv_score.mean(),
        'cv_std': cv_score.std(),
        'precision': class_report['1']['precision'],
        'recall': class_report['1']['recall'],
        'f1_score': class_report['1']['f1-score']
    }
```

## model performance coparison

```python
results_df = pd.DataFrame(results).T
results_df = results_df[['test_auc','test_accuracy','precision','recall','f1_score','cv_mean']]
results_df.columns = ['Accuracy', 'AUC Score', 'Precision', 'Recall', 'F1-Score', 'CV Score']
results_df = results_df.sort_values('AUC Score', ascending=False)

```

```python
display(results_df.style.background_gradient(cmap='Blues',
                                             subset=['AUC Score', 'F1-Score']))
```

```python
best_model_name = results_df.index[0]
best_model = best_models[best_model_name]
```

```python
print(f"\n BEST MODEL: {best_model_name}")
print(f" Best AUC Score: {results_df.loc[best_model_name, 'AUC Score']:.4f}")
print(f" Best F1-Score: {results_df.loc[best_model_name, 'F1-Score']:.4f}")
```

```python
# Cell 15: Best Model Detailed Analysis
print(" Detailed Analysis of Best Model")

# Get predictions from best model
y_pred_best = best_model.predict(X_test)
y_pred_proba_best = best_model.predict_proba(X_test)[:, 1]

# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred_best,
    'Predicted_Probability': y_pred_proba_best
})
comparison_df['Correct'] = comparison_df['Actual'] == comparison_df['Predicted']

print(" Prediction Summary (First 10 samples):")
display(comparison_df.head(10))

# Calculate accuracy and auc for best model
best_accuracy = accuracy_score(y_test, y_pred_best)
best_auc = roc_auc_score(y_test, y_pred_proba_best)
best_precision = precision_score(y_test, y_pred_best)
best_recall = recall_score(y_test, y_pred_best)
best_f1 = f1_score(y_test, y_pred_best)

print(f"\n Overall Performance:")
print(f"Accuracy: {best_accuracy:.4f}")
print(f"AUC Score: {best_auc:.4f}")
print(f"Precision: {best_precision:.4f}")
print(f"Recall: {best_recall:.4f}")
print(f"F1-Score: {best_f1:.4f}")
```

```python
# Confusion Matrix
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Churn', 'Churn'], 
            yticklabels=['No Churn', 'Churn'])
plt.title(f'Confusion Matrix\n({best_model_name})')
plt.ylabel('Actual')
plt.xlabel('Predicted')
```

```python
# ROC Curve
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 2)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_best)
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {best_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
```

```python
# Feature Importance (if available)
plt.subplot(1, 3, 3)
if hasattr(best_model.named_steps['model'], 'feature_importances_'):
    feature_importance = best_model.named_steps['model'].feature_importances_
    feature_names = X.columns
    
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': feature_importance
    }).sort_values('importance', ascending=False).head(10)
    
    sns.barplot(data=importance_df, x='importance', y='feature', palette='viridis')
    plt.title('Top 10 Feature Importance')
else:
    # Show correlation with target as alternative
    corr_with_target = pd.DataFrame({
        'feature': X.columns,
        'correlation': [np.corrcoef(X[col], y)[0, 1] for col in X.columns]
    }).sort_values('correlation', key=abs, ascending=False).head(10)
    
    sns.barplot(data=corr_with_target, x='correlation', y='feature', palette='rocket')
    plt.title('Top 10 Features Correlated with Churn')

plt.tight_layout()
plt.show()
```
